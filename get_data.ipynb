```python
import re
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup, SoupStrainer
import urllib
import os
from argparse import Namespace
from urllib.request import urlopen
import time
```

```python
url = "https://dergipark.org.tr/tr/search/3?section=articles"
strainer = SoupStrainer('div', class_='article-cards')
page = requests.get(url, allow_redirects=False, timeout=100)
soup = BeautifulSoup(page.text, features="lxml", parse_only=strainer)
print(soup.prettify)
```

```python
import random as random
import sys

b = {}
score = 0

while True:
    try:
        a = input("Level: ")
        if int(a) == 1 or int(a) == 2 or int(a) == 3:
            if int(a) == 1:
                x = 0
                y = 0
                while x in range(10):
                    c = random.choice(range(1, 10))
                    d = random.choice(range(1, 10))
                    b[f"{c} + {d} = "] = c + d
                    x += 1
                f = []
                for z in b.keys():
                    f.append(z)

                n = 0
                while n < 10:
                    try:
                        g = int(input(f"{f[n]}"))
                        if g == b[f[n]]:
                            n += 1
                            score += 1
                        else:
                            y += 1
                            if y == 3:
                                print(f"{f[n]} {b[f[n]]}")
                                n += 1
                            else:
                                print("EEE")
                    except ValueError:
                        y += 1
                        if y == 3:
                            print(f"{f[n]} {b[f[n]]}")
                            n += 1
                        else:
                            print("EEE")

                print(score)
                sys.exit()
            elif int(a) == 2:
                x = 0
                y = 0
                while x in range(10):
                    c = random.choice(range(10, 100))
                    d = random.choice(range(10, 100))
                    b[f"{c} + {d} = "] = c + d
                    x += 1
                f = []
                for z in b.keys():
                    f.append(z)

                n = 0
                while n < 10:
                    try:
                        g = int(input(f"{f[n]}"))
                        if g == b[f[n]]:
                            n += 1
                            score += 1
                        else:
                            y += 1
                            if y == 3:
                                print(f"{f[n]} {b[f[n]]}")
                                n += 1
                            else:
                                print("EEE")
                    except ValueError:
                        y += 1
                        if y == 3:
                            print(f"{f[n]} {b[f[n]]}")
                            n += 1
                        else:
                            print("EEE")

                print(score)
                sys.exit()

            else:
                x = 0
                y = 0
                while x in range(10):
                    c = random.choice(range(100, 1000))
                    d = random.choice(range(100, 1000))
                    b[f"{c} + {d} = "] = c + d
                    x += 1
                f = []
                for z in b.keys():
                    f.append(z)

                n = 0
                while n < 10:
                    try:
                        g = int(input(f"{f[n]}"))
                        if g == b[f[n]]:
                            n += 1
                            score += 1
                        else:
                            y += 1
                            if y == 3:
                                print(f"{f[n]} {b[f[n]]}")
                                n += 1
                            else:
                                print("EEE")
                    except ValueError:
                        y += 1
                        if y == 3:
                            print(f"{f[n]} {b[f[n]]}")
                            n += 1
                        else:
                            print("EEE")

                print(score)
                sys.exit()
    except ValueError:
        continue
```
